```python
#pandas 라이브러리를 사용하기 앞서 pd라 정의
import pandas as pd
# 데이터셋을 읽기
df = pd.read_csv ("data/iris.csv")
# 데이터셋의 컬럼 이름 출력
print(df.columns)
# 데이터셋의 처음 3 개 항목 출력 , n개 가능
print(df.head(3))
```

```python
# 평균
print(df["Sepal.Length"].mean())
# 분산
print(df["Sepal.Length"].var())
# 표준편차
print(df[u'Sepal.Length'].std())
# 비대칭도
print(df["Sepal.Length"].skew())
# 첨도
print(df["Sepal.Length"].kurtosis())
```

```python
# 이미지 불러오는 라이브러리 scipy 정의
import scipy.misc
# scipy.misc.imread 는 1.2 버전에서 사라지기 때문에 imageio 정의
import imageio
# image 보여주는 라이브러리 정의
from matplotlib import pyplot as plt
# 실행한 브라우저에서 바로 그림을 볼수있게 하기 위한 명령어
%matplotlib inline
# scipy.misc.imread 는 1.2 버전에서 사라지기 때문에 imageio.imread 사용
#testimg = scipy.misc.imread("data/blue_jay.jpg")
testimg = imageio.imread("data/blue_jay.jpg")
# plt를 사용하여 imageio.imread 로 불러온 이미지 출력
plt.imshow( testimg)
# 해당 이미지의 픽셀의 빨강, 녹색, 파랑 채널값을 보여줌
testimg.shape
```

```python
plt.subplot(131)
plt.imshow( testimg[:,:,0], cmap="Reds")
plt.title("Red channel")

plt.subplot(132)
plt.imshow( testimg[:,:,1], cmap="Greens")
plt.title("Green channel")

plt.subplot(133)
plt.imshow( testimg[:,:,2], cmap="Blues")
plt.title("Blue channel")
```
